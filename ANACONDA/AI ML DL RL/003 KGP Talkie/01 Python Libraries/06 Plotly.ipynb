```python
df = pd.DataFrame(np.random.randn(100,3),columns = ['A','B','C'])
df.head()
```

```python
df.iplot()
```

```python
df = pd.DataFrame(np.random.randn(100,3),columns = ['A','B','C'])
df['A'] = df['A'].cumsum()
df['B'] = df['B'].cumsum()
df['C'] = df['C'].cumsum()
df.iplot()
```

```python
plt.plot(df)
```

```python
df.iplot(x='A',y='B',mode='markers',size=16)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['sex'].value_counts()
```

```python
titanic.iplot(kind='bar',x='sex',y='survived',title='Survived',xTitle='Sex',yTitle='#Sruvived')
```

```python
df = pd.DataFrame(np.random.randn(100,3),columns = ['A','B','C'])
df['A'] = df['A'].cumsum() + 20 # To avoid negative areas 
df['B'] = df['B'].cumsum() + 20 # To avoid negative areas
df['C'] = df['C'].cumsum() + 20 # To avoid negative areas
df.iplot()
```

```python
df.iplot(kind='area',fill=True)
```

```python
df3 = pd.DataFrame({'X' :[10,20,30,20,10] , 'Y' : [10,20,30,20,10] , 'Z' : [10,20,30,20,10]})
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
# help(cf.datagen)
cf.datagen.sinwave(10,0.25).iplot(kind='surface')
```

```python
cf.datagen.scatter3d(2,150,mode='stocks').iplot(kind='scatter3d',x='x',y='y',z='z')
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='hist',bins=25,barmode='group',bargap=None)
```

```python
cf.datagen.bubble3d(5,4,mode='stocks').iplot(kind='bubble3d',x='x',y='y',z='z',size='size')
```

```python
cf.datagen.heatmap(20,20).iplot(kind='heatmap',colorscale='spectral',title='Cufflinks - Heatmap')
```
